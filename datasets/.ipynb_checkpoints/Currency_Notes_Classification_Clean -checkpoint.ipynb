```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from pathlib import Path
import warnings
from PIL import Image
import json
from datetime import datetime

warnings.filterwarnings('ignore')


sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 11

print("Libraries imported successfully")
print(f"Analysis Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

## Dataset Loading & Initial Inspection

```python
DATA_PATH = 'Currency_data/USD'
OUTPUT_DIR = '/kaggle/working/'

Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)

print(f"Data Path: {DATA_PATH}")
print(f"Output Directory: {OUTPUT_DIR}")
```

```python
def load_dataset_info():
    records = []
    
    for img_path in Path(DATA_PATH).rglob("*"):
        if img_path.suffix.lower() in ['.jpg', '.jpeg', '.png']:
            records.append({
                "image_path": str(img_path),
                "currency": img_path.parent.name.upper(),
                "filename": img_path.name,
                "file_size_kb": img_path.stat().st_size / 1024
            })
    
    return pd.DataFrame(records)

df_raw = load_dataset_info()

print("="*60)
print("DATASET OVERVIEW")
print("="*60)
print(f"Total Images: {len(df_raw):,}")
print(f"Unique Classes: {df_raw['currency'].nunique()}")
print(f"\nClass Distribution:")
print(df_raw['currency'].value_counts())
print(f"\nFile Size Statistics (KB):")
print(df_raw['file_size_kb'].describe())
```

```python
def display_sample_images(df, n_samples=3):
    classes = df['currency'].unique()
    n_classes = len(classes)
    
    fig, axes = plt.subplots(n_classes, n_samples, figsize=(15, 3*n_classes))
    fig.suptitle('Sample Images from Each Currency Class', fontsize=16, y=1.00)
    
    for i, currency in enumerate(classes):
        samples = df[df['currency'] == currency].sample(min(n_samples, len(df[df['currency'] == currency])))
        
        for j, (_, row) in enumerate(samples.iterrows()):
            img = cv2.imread(row['image_path'])
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            ax = axes[i, j] if n_classes > 1 else axes[j]
            ax.imshow(img)
            ax.set_title(f"{currency}\n{img.shape[1]}x{img.shape[0]}")
            ax.axis('off')
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/sample_images.png', dpi=150, bbox_inches='tight')
    plt.show()

display_sample_images(df_raw)
```

##  Data Cleaning

```python
def remove_corrupt_images(df):
    valid_indices = []
    corrupt_count = 0
    
    print("Checking for corrupt images...")
    
    for idx, row in df.iterrows():
        try:
            with Image.open(row['image_path']) as img:
                img.verify()
            
            test_img = cv2.imread(row['image_path'])
            if test_img is not None:
                valid_indices.append(idx)
            else:
                corrupt_count += 1
        except Exception as e:
            corrupt_count += 1
            print(f"Corrupt: {row['filename']} - {str(e)[:50]}")
    
    print(f"\nValid images: {len(valid_indices)}")
    print(f"Corrupt images removed: {corrupt_count}")
    
    return df.loc[valid_indices].reset_index(drop=True)

df_clean = remove_corrupt_images(df_raw)
```

```python
def extract_image_properties(df):
    properties = []
    
    print("Extracting image properties...")
    
    for _, row in df.iterrows():
        img = cv2.imread(row['image_path'])
        h, w = img.shape[:2]
        
        properties.append({
            'width': w,
            'height': h,
            'aspect_ratio': w / h,
            'total_pixels': w * h,
            'channels': img.shape[2] if len(img.shape) == 3 else 1
        })
    
    return pd.concat([df, pd.DataFrame(properties)], axis=1)

df_clean = extract_image_properties(df_clean)

print("\nImage Property Statistics:")
print(df_clean[['width', 'height', 'aspect_ratio', 'total_pixels']].describe())
```

```python
def filter_quality_issues(df, min_pixels=10000, max_aspect=5, min_aspect=0.2):
    initial_count = len(df)
    
    df = df[df['total_pixels'] >= min_pixels]
    df = df[(df['aspect_ratio'] >= min_aspect) & (df['aspect_ratio'] <= max_aspect)]
    
    removed = initial_count - len(df)
    
    print(f"\nQuality Filtering:")
    print(f"  Removed {removed} low-quality images")
    print(f"  Remaining images: {len(df)}")
    
    return df.reset_index(drop=True)

df_clean = filter_quality_issues(df_clean)
```

```python
def analyze_class_balance(df):
    class_counts = df['currency'].value_counts()
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    class_counts.plot(kind='bar', ax=axes[0], color='steelblue')
    axes[0].set_title('Class Distribution', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Currency Denomination')
    axes[0].set_ylabel('Number of Images')
    axes[0].grid(axis='y', alpha=0.3)
    
    class_counts.plot(kind='pie', ax=axes[1], autopct='%1.1f%%', startangle=90)
    axes[1].set_title('Class Distribution (%)', fontsize=14, fontweight='bold')
    axes[1].set_ylabel('')
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/class_distribution.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    max_class = class_counts.max()
    min_class = class_counts.min()
    imbalance_ratio = max_class / min_class
    
    print("\nClass Balance Analysis:")
    print(f"  Most frequent class: {class_counts.idxmax()} ({max_class} images)")
    print(f"  Least frequent class: {class_counts.idxmin()} ({min_class} images)")
    print(f"  Imbalance ratio: {imbalance_ratio:.2f}")
    
    if imbalance_ratio > 3:
        print("  WARNING: Significant class imbalance detected!")
        print("  Recommendation: Consider data augmentation or weighted loss")
    else:
        print("  Class distribution is reasonably balanced")
    
    return class_counts

class_distribution = analyze_class_balance(df_clean)
```

```python
def remove_rare_classes(df, min_samples=2):
    class_counts = df['currency'].value_counts()
    valid_classes = class_counts[class_counts >= min_samples].index
    rare_classes = class_counts[class_counts < min_samples].index
    
    if len(rare_classes) > 0:
        print(f"\nRemoving {len(rare_classes)} rare classes:")
        for cls in rare_classes:
            print(f"  {cls}: {class_counts[cls]} samples")
        
        df = df[df['currency'].isin(valid_classes)].reset_index(drop=True)
    else:
        print("\nNo rare classes found")
    
    return df

df_clean = remove_rare_classes(df_clean)

print(f"\nFinal cleaned dataset: {len(df_clean)} images across {df_clean['currency'].nunique()} classes")
```

##  Feature Engineering
Currency notes have distinctive visual characteristics:
1. **Color Distribution:** Different denominations use different color schemes
2. **Brightness/Contrast:** Varies across denominations
3. **Edge Patterns:** Security features create unique edge patterns
4. **Texture:** Fine details and patterns differ by denomination

```python
def extract_image_features(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    features = {}
    
    features['brightness_mean'] = np.mean(gray)
    features['brightness_std'] = np.std(gray)
    features['contrast'] = gray.max() - gray.min()
    
    for i, color in enumerate(['blue', 'green', 'red']):
        hist = cv2.calcHist([img], [i], None, [32], [0, 256])
        features[f'{color}_mean'] = np.mean(hist)
        features[f'{color}_std'] = np.std(hist)
        features[f'{color}_max'] = np.max(hist)
    
    edges = cv2.Canny(gray, 50, 150)
    features['edge_density'] = np.sum(edges > 0) / edges.size
    features['edge_mean'] = np.mean(edges)
    
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    features['texture_variance'] = laplacian.var()
    
    b, g, r = cv2.split(img)
    features['blue_dominance'] = np.mean(b) / (np.mean(img) + 1e-6)
    features['green_dominance'] = np.mean(g) / (np.mean(img) + 1e-6)
    features['red_dominance'] = np.mean(r) / (np.mean(img) + 1e-6)
    
    return features
```

```python
print("Extracting features from all images...")
print("This may take a few minutes...\n")

feature_list = []

for idx, row in df_clean.iterrows():
    if idx % 100 == 0:
        print(f"  Progress: {idx}/{len(df_clean)} images processed...", end='\r')
    
    features = extract_image_features(row['image_path'])
    features['currency'] = row['currency']
    features['image_path'] = row['image_path']
    feature_list.append(features)

df_features = pd.DataFrame(feature_list)

print(f"\nFeature extraction complete!")
print(f"\nExtracted Features: {len(df_features.columns) - 2} features per image")
print(f"\nFeature names: {[col for col in df_features.columns if col not in ['currency', 'image_path']]}")
```

```python
df_features.to_csv(f'{OUTPUT_DIR}/extracted_features.csv', index=False)
print(f"Features saved to: {OUTPUT_DIR}/extracted_features.csv")

feature_cols = [col for col in df_features.columns if col not in ['currency', 'image_path']]
print("\nFeature Statistics:")
print(df_features[feature_cols].describe())
```

##  Data Transformation
1. **Label Encoding:** Convert currency names to numeric labels
2. **Feature Scaling:** Standardize features to zero mean and unit variance
3. **Train/Validation/Test Split:** 70/15/15 stratified split

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

le = LabelEncoder()
df_features['label'] = le.fit_transform(df_features['currency'])

print("Label Encoding:")
for i, currency in enumerate(le.classes_):
    print(f"  {currency} -> {i}")

label_mapping = {i: currency for i, currency in enumerate(le.classes_)}
with open(f'{OUTPUT_DIR}/label_mapping.json', 'w') as f:
    json.dump(label_mapping, f, indent=2)

print("\nLabel mapping saved")
```

```python
train_val_df, test_df = train_test_split(
    df_features,
    test_size=0.15,
    stratify=df_features['currency'],
    random_state=42
)

train_df, val_df = train_test_split(
    train_val_df,
    test_size=0.176,
    stratify=train_val_df['currency'],
    random_state=42
)

print("Dataset Split:")
print(f"  Training set: {len(train_df)} images ({len(train_df)/len(df_features)*100:.1f}%)")
print(f"  Validation set: {len(val_df)} images ({len(val_df)/len(df_features)*100:.1f}%)")
print(f"  Test set: {len(test_df)} images ({len(test_df)/len(df_features)*100:.1f}%)")

print("\nClass distribution per split:")
print("\nTraining:")
print(train_df['currency'].value_counts())
print("\nValidation:")
print(val_df['currency'].value_counts())
print("\nTest:")
print(test_df['currency'].value_counts())
```

```python
train_df.to_csv(f'{OUTPUT_DIR}/train_set.csv', index=False)
val_df.to_csv(f'{OUTPUT_DIR}/val_set.csv', index=False)
test_df.to_csv(f'{OUTPUT_DIR}/test_set.csv', index=False)

print("Train/Val/Test splits saved")
print("\nDATA PREPROCESSING COMPLETED SUCCESSFULLY!")
```

## Statistical Analysis

```python
feature_cols = [col for col in df_features.columns if col not in ['currency', 'image_path', 'label']]

correlation_matrix = df_features[feature_cols].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=0.5)
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

print("Highly Correlated Features (|r| > 0.8):")
high_corr = np.where(np.abs(correlation_matrix) > 0.8)
high_corr_pairs = [(correlation_matrix.index[x], correlation_matrix.columns[y], correlation_matrix.iloc[x, y]) 
                   for x, y in zip(*high_corr) if x != y and x < y]

for feat1, feat2, corr in high_corr_pairs:
    print(f"  {feat1} <-> {feat2}: {corr:.3f}")
```

```python
print("\nSummary Statistics by Currency Class:\n")

for currency in df_features['currency'].unique():
    print(f"\n{'='*60}")
    print(f"Currency: {currency}")
    print(f"{'='*60}")
    
    subset = df_features[df_features['currency'] == currency][feature_cols]
    print(subset.describe().loc[['mean', 'std', 'min', 'max']].T)
```

```python
important_features = ['brightness_mean', 'edge_density', 'red_mean', 'currency']
pair_plot_data = df_features[important_features]

pairplot = sns.pairplot(pair_plot_data, hue='currency', diag_kind='kde', 
                        plot_kws={'alpha': 0.6}, height=3)
pairplot.fig.suptitle('Pair Plot: Key Features vs Currency Classes', 
                      fontsize=16, fontweight='bold', y=1.01)
plt.savefig(f'{OUTPUT_DIR}/feature_pairplot.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = df_features[feature_cols].values
y = df_features['currency'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(12, 8))
for currency in df_features['currency'].unique():
    mask = y == currency
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1], label=currency, alpha=0.6, s=50)

plt.xlabel(f'First Principal Component ({pca.explained_variance_ratio_[0]*100:.1f}% variance)', fontsize=12)
plt.ylabel(f'Second Principal Component ({pca.explained_variance_ratio_[1]*100:.1f}% variance)', fontsize=12)
plt.title('PCA Projection of Currency Notes', fontsize=16, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.savefig(f'{OUTPUT_DIR}/pca_visualization.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nTotal variance explained by 2 components: {sum(pca.explained_variance_ratio_)*100:.2f}%")
```
